U okviru ovog projektnog zadatka, koji je radjen kao projektni zadatak za kurs Efemeridska astronomija 1, prikazuju se rezultati Keplerove jednačine, dobijeni u radu "The dynamics of Kepler equation", V. Lanchares i I. L. Pérez Barrón,2003.

Keplerova jednačina se koristi za opisivanje putanje planeta oko Sunca. Omogućava određivanje položaja planete u orbiti u određenom trenutku i omogućava nam da matematički objasnimo kretanja nebeskih tela. 

Keplerova jednačina se definiše:
$$ E = M + e * sin(E) $$

Gde je:
- E - ekscentrična anomalija
- M - srednja anomalija
- e - ekscentricitet

Keplerova jednačina je jedna od najpoznatijih nelinearnih jednačina. Postoje različite metode za rešavanje Keplerove jednačine poput: numeričke metode, serije rešenja, numeričke integracije, geometrijske metode...
Neke od najpoznatijih metoda za rešavanje nelinearnih jednačina su metoda polovljenja intervala(binarna pretraga), metoda iterativnog poboljšanja (Newton-Raphson metoda, metoda sečice i metoda fiksne tačke),metode bazirane na konačnim razlikama, metoda regula-falsi, metoda optimizacije i druge.

Prilikom rešavanja Keplerove nelinearne jednačine, koristiće se iterativna metoda za rešavanje nelinearne jednačine:
$$ E_{n+1}= M + e*sin(E_n)$$

Ekscentricitet je pojam koji se koristi kako bi se opisala karakteristika elipse ili orbite oko fokusa. Ekscentricitet je numerička vrednost koja meri "spuštenost" ili "nepravilnost" oblika elipse u odnosu na kružnicu.
Ekscentricitet je uvek vrednost između 0 i 1. Ako je ekscentricitet jednak nuli, to znači da je elipsa zapravo kružnica (kada su oba fokusa u centru elipse). Kada je ekscentricitet jednak 1, to znači da je elipsa izdužena u jednom pravcu do tačke da postaje parabola, a kada je ekscentricitet veći od 1, elipsa postaje hiperbola.

U radu su prikazane vrednosti dobijene nakon različitog broja iteracija i za različite početne uslove.

Keplerova jednačina, koja matematički definiše odnos između ekscentrične anomalije (E), srednje anomalije (M) i ekscentriciteta (e), ima široku primenu u astrofizici i orbitalnoj mehanici.

Keplerovu jednačinu predstavljamo pomoću formule:
$$ F^{n}(e,E_0,M) = M + e*sin(E) $$
gde je n broj iteracija, M srednja anomalija, e ekscentricitet i E eskcentrična anomalija.

```python
import cmath
import numpy as np
import matplotlib.pyplot as plt
import astropy as ap
import pandas as pd
import seaborn as sns
```

## $$F^{1}(e,E_0,M)$$

početni uslovi: E=0, broj iteracija = 1

```python
def keplerova_jednacina(e,E,M):
    E=M
    for i in range(1):
        E = M + e*cmath.sin(E)

    return E

M_osa = np.linspace(-4,4,600)
e_osa = np.linspace(-3,3,600)

E_mat = np.empty((len(M_osa),len(e_osa),))
e_duzina = len(e_osa)
M_duzina = len(M_osa)

for i in range(e_duzina):
    for j in range(M_duzina):
        E = keplerova_jednacina(e_osa[i],0,M_osa[j])
        E_mat[i][j] = np.abs(E)

M,e = np.meshgrid(M_osa,e_osa)

plt.figure(dpi=100)
plt.pcolormesh(M_osa,e_osa,E_mat,cmap='viridis')
plt.colorbar()

plt.title('Keplerova jednacina')
plt.xlabel('M')
plt.ylabel('e')
plt.show()
```

Grafik prikazuje vrednosti srednje anomalije(E) nakon jedne iteracije, sa početnim uslovom E = 0, za različite uredjene parove (M,e).

Iz datog grafika, možemo da vidimo simetriju u odnosu na y-osu, M=0. Ona će se pojaviti jer u Keplerovoj jednačini imamo periodičnu funkciju sin(E).
