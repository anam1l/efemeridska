U okviru ovog projektnog zadatka, koji je radjen kao projektni zadatak za kurs Efemeridska astronomija 1, prikazuju se rezultati Keplerove jednačine, dobijeni u radu "The dynamics of Kepler equation", V. Lanchares i I. L. Pérez Barrón,2003.

Keplerova jednačina se koristi za opisivanje putanje planeta oko Sunca. Omogućava određivanje položaja planete u orbiti u određenom trenutku i omogućava nam da matematički objasnimo kretanja nebeskih tela. 

Keplerova jednačina se definiše:
$$ E = M + e * sin(E) $$

Gde je:
- E - ekscentrična anomalija
- M - srednja anomalija
- e - ekscentricitet

U radu su prikazane vrednosti dobijene nakon različitog broja iteracija i za različite početne uslove.

Keplerovu jednačinu predstavljamo pomoću formule:
$$ F^{n}(e,E_0,M) = M + e*sin(E) $$
gde je n broj iteracija, M srednja anomalija, e ekscentricitet i E eskcentrična anomalija.

Ukoliko odlučimo da rešenja Keplerove jednačine predstavimo na kompleksnoj ravni, možemo prikazati Julijeve skupove.Julijevi skupovi definisani su kao:
$$ J(M,e) = \{z \in C | {F^{n}(M,e;z)}_{n\in \mathbb{N}} \space bounded \} $$

Gaston Julia set je geometrijski skup u kompleksnoj ravni, nazvan po francuskom matematičaru Gastonu Juliji. Ovi skupovi su posebni tipovi fraktala koji se stvaraju primenom iterativnih procesa na kompleksnim brojevima.

Julijev set se konstruiše iterativno primenjujući određenu funkciju na svaki kompleksni broj u ravni i posmatrajući ponašanje tih iteracija. Osnovni proces u stvaranju Julijevog skupa uključuje odabir početne tačke u kompleksnoj ravni i primenu iterativne funkcije. Na osnovu toga se određuje hoće li tačka konvergirati ka određenom skupu ili divergirati.

Julijevi skupovi se vizualizuju bojenjem kompleksne ravni na osnovu toga da li iteracije kompleksnih brojeva iz jednog skupa konvergiraju ili divergiraju. Ovi skupovi su značajni u matematici i kompleksnoj analizi zbog svoje lepe vizuelne strukture i kompleksnih oblika koji se formiraju. Oni omogućavaju dublje razumevanje iterativnih procesa i ponašanja kompleksnih brojeva.
Iz grafičkih prikaza Julijevih skupova možemo izvesti nekoliko važnih karakteristika: konvergenciju i divergenciju funkcije, granice funkcije i njegnu fraktalnu strkuturu, stabilnost iterativnih procesa kod nekih funkcija, topološke osobine i drugo.

```python
import numpy as np
import matplotlib.pyplot as plt

def julia_skup(xmin, xmax, ymin, ymax, sirina, visina, max_iter, M, e):
    x = np.linspace(xmin, xmax, sirina)
    y = np.linspace(ymin, ymax, visina)
    X, Y = np.meshgrid(x, y)
    Z = X + 1j * Y
    matrica = np.zeros((visina, sirina))

    for i in range(max_iter):
        Z = M + e * np.sin(Z)
        uslov = np.abs(Z) < 1000
        matrica += uslov

    #matrica = np.log(matrica + 1)  # bolja slika
    plt.imshow(matrica, extent=(xmin, xmax, ymin, ymax), cmap='hot')
    plt.title('Julijev skup \n M=0.8, e = 0.1 + i')
    plt.xlabel('Realni deo')
    plt.ylabel('Imaginarni deo')
    plt.colorbar()
    plt.show()

# parametri
xmin, xmax = -6, 6
ymin, ymax = -6, 6
sirina, visina = 800, 800
max_iter = 100

# definisanje varijabla
M = complex(0.8, 0)
e = complex(0.1,1)

julia_skup(xmin, xmax, ymin, ymax, sirina, visina, max_iter, M, e)
```
